# **Comparing age and experience levels among teams**

```python
team_summary = df.groupby('2024 Team').agg({
    'Age': 'mean',
    'Test caps': 'sum',
    'ODI caps': 'sum',
    'T20 caps': 'sum'
}).reset_index()
team_summary.columns = ['Team', 'Average Age', 'Total Test Caps', 'Total ODI Caps', 'Total T20 Caps']
print(team_summary)
```

```python
team_summary=team_summary.sort_values(by='Average Age', ascending=False)
print(team_summary)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=team_summary, x='Team', y='Average Age', palette='coolwarm')
plt.title('Average Age of Players by Team', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Average Age', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.show()
```

```python
experience_melted = team_summary.melt(
    id_vars='Team',
    value_vars=['Total Test Caps', 'Total ODI Caps', 'Total T20 Caps'],
    var_name='Experience Type',
    value_name='Caps'
)


plt.figure(figsize=(14, 8))
sns.barplot(data=experience_melted, x='Team', y='Caps', hue='Experience Type', palette='viridis')
plt.title('Experience Levels (Caps) by Team', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Total Caps', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Experience Type')
plt.show()
```

# **Trends in reserve pricing for capped vs. uncapped players**

# **Uncapped players who played ipl 2024**

```python
# Ensure 2024 IPL column has consistent NaN values for blanks or nulls
df['2024\nIPL'] = df['2024\nIPL'].replace(['', ' ', 'NA', 'None'], pd.NA)

# Filter for uncapped players
uncapped_players = df[df['C/U/A'] == 'Uncapped']

# Filter uncapped players with non-zero, non-blank, non-NaN values in '2024 IPL' column
valid_uncapped_players = uncapped_players[
    (uncapped_players['2024\nIPL'].notna()) &  # Non-NaN or non-null
    (uncapped_players['2024\nIPL'] != 0)      # Non-zero
]

valid_uncapped_players_sorted = valid_uncapped_players.sort_values(by='2024\nIPL', ascending=False)

# Display the results
print("Uncapped Players with Valid '2024 IPL' Values:")
print(valid_uncapped_players_sorted[['First Name', 'Surname', '2024\nIPL', 'Specialism']])
```

# **Uncapped Players Who Played IPL**

```python
# Replace placeholders like blanks or strings with NaN for consistent handling
df['2024\nIPL'] = df['2024\nIPL'].replace(['', ' ', 'NA', 'None'], pd.NA)
df['IPL'] = df['IPL'].replace(['', ' ', 'NA', 'None'], pd.NA)

# Filter for uncapped players
uncapped_players = df[df['C/U/A'] == 'Uncapped']

# Apply the conditions to filter players who played IPL in both columns
valid_uncapped_players = uncapped_players[
    (uncapped_players['2024\nIPL'].notna()) & (uncapped_players['2024\nIPL'] != 0) &  # Non-NaN and non-zero for 2024 IPL
    (uncapped_players['IPL'].notna()) & (uncapped_players['IPL'] != 0)             # Non-NaN and non-zero for IPL
]

# Calculate the total by adding 'IPL' and '2024 IPL' columns
valid_uncapped_players['Total IPL'] = valid_uncapped_players['2024\nIPL'] + valid_uncapped_players['IPL']

# Sort the players in descending order based on the total
sorted_uncapped_players = valid_uncapped_players.sort_values(by='Total IPL', ascending=False)

# Display the results
print("Uncapped Players Who Played IPL (Sorted by Total IPL):")
print(sorted_uncapped_players[['First Name', 'Surname', 'IPL', '2024\nIPL', 'Total IPL']])
```
